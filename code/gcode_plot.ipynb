```python

# Initialize empty lists for X and Y coordinates
x_coords = []
y_coords = []

# Split the text into lines
lines = text.strip().split("\n")

# Extract X and Y coordinates from each line
for line in lines:
    parts = line.split()
    for part in parts:
        if part.startswith("X"):
            x_coords.append(float(part[1:]))  # Extract the numeric value after X
        elif part.startswith("Y"):
            y_coords.append(float(part[1:]))  # Extract the numeric value after Y

# Create NumPy arrays
a = np.array(x_coords)
b = np.array(y_coords)

print("Array a (X coordinates):", a)
print("Array b (Y coordinates):", b)
```

```python
pt.plot(a,b,'-ob',linewidth=7)
```

```python
pluse=0
for i in range(np.size(a)-1):
    hh=np.sqrt((a[i+1]-a[i])**2+((b[i+1]-b[i])**2))
    pluse=pluse+hh
```
